```python
df= pd.read_csv('Sample.csv')
df.head()
```

```python
df= df.tail()
df
```

```python
df= df.info()
df
```

```python
df['location']= df['latitude'].astype(str) + ',' + df['longitude'].astype(str)
df.head()
```

```python
df['team']= df['cohort'].astype(str) + '_' + df['cycle'].astype(str)
df.head()
```

```python
df= df.drop(['latitude','longitude','cohort','cycle'], axis=1)
df.head()
```

```python
df= df.describe()
df
```

```python
import matplotlib.pyplot as plt

gender_counts = df["gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
marital_counts = df["marital_status"].value_counts().sort_index()

plt.figure(figsize=(7,7))
plt.pie(
    marital_counts,
    labels=marital_counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Marital Status Distribution")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['latitude'], df['longitude'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Distribution of farmers for the organization')
plt.show
```
